# Introduccio a la base de dades

La base de dades que analitzaré en aquest Notebook s'anomena Dades de Rendiment Corporal.
L'objectiu principal serà classificar la salut d'una persona depenent d'unes dades d'entrada com pot ser l'edat, l'altura, el pes o el percentatge de grassa corporal. 
Hi ha quatre tipus de "salut" en una persona: A, B, C i D, sent A una salut molt bona i D una salut molt dolenta.
La base de dades està composta per 12 atributs i 13393 instàncies.

# Llibreries

- Numpy: dona suport per crear vectors i matrius multidimensionals. Gran col·lecció de funcions matemàtiques d'alt nivell.
- Pandas: ofereix estructures de dades i operaciones per manipular taules numèriques.
- Matplotlib: biblioteca per la generació de gràfiques de dades contingudes en llistes o arrays.
- Seaborn: serveix per generar gràfiques més complexes a partir de dades contingudes en llistes o arrays.
- Sklearn: inclou els models de classificacio, regressió i analisi de grups, a més de totes les operacions necessàries.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Perceptron
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.metrics import classification_report, accuracy_score, f1_score, recall_score, precision_score, precision_recall_curve, average_precision_score, roc_curve, auc
```

# Visualització de les dades

Abans de poder començar a classificar, analitzaré les dades per saber com estan organitzades i quins atributs 
són els més significatius de la base de dades.

```python
def load_dataset(path):
    dataset = pd.read_csv(path, header=0, delimiter=',')
    dataset['gender'] = dataset['gender'].astype('category')
    dataset['gender'] = dataset['gender'].cat.codes
    dataset['class'] = dataset['class'].astype('category')
    dataset['class'] = dataset['class'].cat.codes
    return dataset
dataset = load_dataset("../dataset/bodyPerformance.csv")
dataset.head(8)
```

Els atributs gènere i classe són atributs categòrics. Els he transformat en atributs numèrics per un millor funcionament.
Ara el gènere correspon a 1 (M) i 0 (F). La classe de salut correspon a 0 (A), 1 (B), 2 (C) i 3 (D).

# Transformació de les dades

```python
dataset.describe()
```

Mirant aquestes dades, es pot veure que el rang que prenen tots els atributs és molt diferent, la qual cosa pot fer que els diferents models que apliqui no puguin classificar bé.
Per tal d'ajustar tots els atributs al mateix rang de dades, hauré de normalitzar les dades.

```python
titles = dataset.columns.values
values = dataset.values
x = values[:,:-1]
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

# Models i rendiment

He provat cinc models de classificació: "Support Vector Machines", "KNN", "Decision Tree", "Random Forest", "Logistic Regression".
Per cadascun, he mirat el seu rendiment dividint les dades en 80%, 70% i 50% d'entrenament. També he aplicat el k-fold des de k=2 fins a k=6.
Cal destacar que els models han estat entrenats amb els seus paràmetres per defecte.
Solament mostraré els models amb millor i pitjor resultat.

```python
y = values[:,-1]
title_x = titles
title_x = np.delete(title_x, 11, 0)
title_y = titles[11]
particions = [0.5, 0.8, 0.7]
model = KNeighborsClassifier()
for part in particions:
        x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=part)
        model.fit(x_train, y_train)
        print("Score del model amb ", part, ": ", model.score(x_test, y_test))
for k in range(2,7):
    scores = cross_val_score(model, x, y, cv=k)
    print("Score promig amb k-fold = ", k, " : ", scores.mean())
```

El model amb pitjor rendiment dels cinc provats ha estat el kNN. El seu score oscil·la entre 0.585 i 0.603, que és solament una mica millor que saber el resultat del llançament d'una moneda.

```python
model = RandomForestClassifier()
for part in particions:
        x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=part)
        model.fit(x_train, y_train)
        print("Score del model amb ", part, ": ", model.score(x_test, y_test))
for k in range(2,7):
    scores = cross_val_score(model, x, y, cv=k)
    print("Score promig amb k-fold = ", k, " : ", scores.mean())
```

En canvi, el model que ha donat millors resultats ha estat el Random Forest. Tots els scores oscil·len entre 0.713 i 0.737.
La resta de models tenen un rendiment d'entre un 61% i un 68%. 

He generat les corbes ROC i PR de tots els models. En aquest cas mostraré les corbes de tots els models, per veure com funciona cadascun d'ells. Primer mostraré les corbes PR.

```python
models = [svm.SVC(probability=True), KNeighborsClassifier(), DecisionTreeClassifier(), RandomForestClassifier(), LogisticRegression()]
nom_models = ["Support Vector Machines", "KNN", "Decision Tree", "Random Forest", "Logistic Regression"]
n_clases = dataset["class"].nunique()
sns.set(rc={'figure.figsize':(5,5)})

for i,model in enumerate(models):
    x_t, x_v, y_t, y_v = train_test_split(x, y, train_size=0.8)
    model.fit(x_t,y_t)
    probs = model.predict_proba(x_v)
    precision = {}
    recall = {}
    average_precision = {}
    plt.figure()
    for j in range(n_clases):
        precision[j], recall[j], _ = precision_recall_curve(y_v == j, probs[:, j])
        average_precision[j] = average_precision_score(y_v == j, probs[:, j])
        plt.plot(recall[j], precision[j],
        label='Precision-recall curve of class {0} (area = {1:0.2f})'''.format(j, average_precision[j]))
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title(nom_models[i])
        plt.legend(loc="upper right")
```

Dels totes les corbes PR generades, un altre cop es pot veure com el Random Forest és el model que funciona millor per les dades d'aquesta base de dades. L'àrea de les corbes PR de cada classe oscil·la entre 0.63 i 0.93, sent la seva mitjana de 0.78.
El kNN, que era el model amb pitjors scores, no és el que té les pitjors àrees en la seva corba PR. Aquestes oscil·len entre 0.37 i 0.81, amb mitjana de 0.58. El pitjor resultat ha estat, en aquest cas, el Decision Tree, amb mitjana de 0.51 (L'score obtingut amb les particions i el k-fold ha estat aproximadament d'un 61%, també bastant baix). 
Comparant aquests resultats amb la resta de models provats, es pot veure com destaca per sobre dels demés el Random Forest.

```python
for i,model in enumerate(models):
    x_t, x_v, y_t, y_v = train_test_split(x, y, train_size=0.8)
    model.fit(x_t,y_t)
    probs = model.predict_proba(x_v)
    precision = {}
    recall = {}
    average_precision = {}
    plt.figure()
    fpr = {}
    tpr = {}
    roc_auc = {}
    for j in range(n_clases):
        fpr[j], tpr[j], _ = roc_curve(y_v == j, probs[:, j])
        roc_auc[j] = auc(fpr[j], tpr[j])

    plt.figure()
    rnd_fpr, rnd_tpr, _ = roc_curve(y_v>0, np.zeros(y_v.size))
    plt.plot(rnd_fpr, rnd_tpr, linestyle='--', label='Sense capacitat predictiva')
    for j in range(n_clases):
        plt.plot(fpr[j], tpr[j], label='ROC curve of class {0} (area = {1:0.2f})' ''.format(j, roc_auc[j]))
    plt.title(nom_models[i])
    plt.legend()
```

SVM 0.89 kNN 0.81 Decision Tree 0.76 Random Forest 0.92 Logistic regression 0.85
De totes les corbes ROC generades, segueix sent la millor la corresponent al Random Forest. Les àrees de les diferents classes oscil·len entre 0.86 i 0.96, sent la seva mitjana de 0.92.  
Cal destacar que per les corbes ROC, els altres models han generat resultats bastant bons també, tots entre 0.76 i 0.89. Però el millor en tots els aspectes continua sent el Random Forest.

# Cerca dels millors hiperparàmetres

Per finalitzar amb la visualització del rendiment dels diferents models, he decidit fer una cerca dels millors hiperparàmetres de cada model, mirant el rendiment amb aquests millors paràmetres. Solament mostraré els millors hiperparàmetres i el rendiment del model que pitjor funciona i del que millor funciona.

```python
model = KNeighborsClassifier()
param_knn = { 'n_neighbors' : [5,7,9,11,13,15], 'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}
grid = GridSearchCV(model, param_knn, verbose=3, n_jobs=-1)
grid.fit(x,y)
print("Els millors parametres: ",grid.best_params_)
print("El millor score: ", grid.best_score_)
```

El kNN, que era el model que pitjor score tenia, ha aconseguit amb els millors paràmetres, aproximadamet un 64% d'score. No millora gaire el rendiment respecte els paràmetres per defecte, exactament millora un 6.7%, concluint que no és un bon model per classificar amb les dades d'aquesta base de dades.

```python
model = RandomForestClassifier()
param_randomforest = {'bootstrap': [True], 'max_depth': [80, 90, 100, 110], 'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],'min_samples_split': [8, 10, 12],'n_estimators': [100, 200, 300, 1000]}

grid = GridSearchCV(model, param_randomforest, verbose=3, n_jobs=-1)
grid.fit(x,y)
print("Els millors parametres: ",grid.best_params_)
print("El millor score: ", grid.best_score_)
```

Un cop buscats els millors paràmetres i calculat el seu rendiment, es pot veure que aquest supera per primer cop el 74%. No és un rendiment excessivament bo, però d'entre els cinc provats, és el millor. Els millors hiperparàmetres solament augmenten el rendiment en un 1.6%.
